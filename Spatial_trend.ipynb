```python
#import libraries

import xarray as xr
import matplotlib.pyplot as plt
```

```python
#in this method i used cdo to compute the trend, season and residuals and imported the datasets into python xarray
#read the datasets
WA= xr.open_dataset('West_Afrique_tm.nc')
s = xr.open_dataset('season1.nc')
trend = xr.open_dataset('a.nc')
residual = xr.open_dataset('new1.nc')
s1 = s['AOD_550_Dark_Target_Deep_Blue_Combined_Mean_Mean']
ds1 = ds['AOD_550_Dark_Target_Deep_Blue_Combined_Mean_Mean']
trend1 = trend['AOD_550_Dark_Target_Deep_Blue_Combined_Mean_Mean']
residual1 = residual['AOD_550_Dark_Target_Deep_Blue_Combined_Mean_Mean']
WA1 = WA['AOD_550_Dark_Target_Deep_Blue_Combined_Mean_Mean']
```

```python
#Plotting
t= trend1.mean(dim='lat').plot(size=3)
plt.title('trend')
r= residual1.mean(dim='lat').plot(size=3)
plt.title('residuals')
s= s1.mean(dim = 'lat').plot(size = 3)
plt.title('seasonal')
w = WA1.mean(dim = 'lat').plot(size = 3)
plt.title('original')
```
