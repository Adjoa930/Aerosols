```python
#import libraries

import xarray as xr
import matplotlib.pyplot as plt
import seaborn as sns
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()
sns.set_style('darkgrid')
```

```python
#load datasets
ds1 = xr.open_dataset('West_Africa_aqua2.nc')
ds2 = xr.open_dataset('seasonal.nc')
ds3 = xr.open_dataset('residual.nc')
ds4 = xr.open_dataset('trends.nc')
```

```python
#select the variables
AOD = ds1['AOD_550_Dark_Target_Deep_Blue_Combined_Mean_Mean']
season = ds2['AOD_550_Dark_Target_Deep_Blue_Combined_Mean_Mean']
resid = ds3['AOD_550_Dark_Target_Deep_Blue_Combined_Mean_Mean']
trend = ds4['AOD_550_Dark_Target_Deep_Blue_Combined_Mean_Mean']
```

```python
r = resid.mean(dim =  'lat')
r1 = r.mean(dim = 'time')
tr = trend.mean(dim = 'lat')
tr1 = tr.mean(dim= 'time')
s = season
s = season.mean(dim = 'lat')
s1 = s.mean(dim = 'time')
A = AOD.mean(dim = 'lat')
A1 = A.mean(dim = 'time')
```

```python
fig = plt.figure(figsize=(12,10))
A1.plot(label = 'original')
s1.plot(label = 'season')
r1.plot(label = 'resid')
tr1.plot(label ='trend' )
plt.legend(loc='best')
plt.savefig('trends.png')
```
