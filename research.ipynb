```python
#import libraries

import xarray as xr
import matplotlib.pyplot as plt
import seaborn as sns
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()
sns.set_style('darkgrid')
```

```python
#load datasets
ds1 = xr.open_dataset('West_Africa_aqua2.nc')
ds2 = xr.open_dataset('seasonal.nc')
ds3 = xr.open_dataset('residual.nc')
ds4 = xr.open_dataset('trends.nc')
ds5 = xr.open_dataset('trend_estimate.nc')
```

```python
#select the variables
AOD = ds1['AOD_550_Dark_Target_Deep_Blue_Combined_Mean_Mean']
season = ds2['AOD_550_Dark_Target_Deep_Blue_Combined_Mean_Mean']
resid = ds3['AOD_550_Dark_Target_Deep_Blue_Combined_Mean_Mean']
trend = ds4['AOD_550_Dark_Target_Deep_Blue_Combined_Mean_Mean']
inter = ds5['AOD_550_Dark_Target_Deep_Blue_Combined_Mean_Mean']
```

```python
seas = season.mean(dim=('lon','lat'))
orig = AOD.mean(dim=('lon','lat'))
res = resid.mean(dim=('lon','lat'))
tre = inter.mean(dim=('lon','lat'))
```

```python
fig = plt.figure(figsize=(12,10))
seas.plot(label = 'seasons')
orig.plot(color = 'red',label = 'AOD')
res.plot(label = 'residuals', color= 'purple')
tre.plot(color ='black', label = 'trends')
plt.legend(loc='best')
plt.savefig('plot1.png')
```
